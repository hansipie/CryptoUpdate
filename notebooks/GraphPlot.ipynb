# Graph crypto

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def rapport(tokenlist, all):
    df=pd.DataFrame(index=all.index)
    for t in tokenlist:
        df[t]=all[t]
    return df
```

Transformation des fichiers CSV en DataFrame

```python
all = pd.read_csv("../outputs/ArchiveFinal.csv", index_col="Timestamp")
all = all.sort_index().reindex(sorted(all.columns), axis=1)
all.tail()
```

## Pie Chart

```python
# read csv file
all = pd.read_csv("ArchiveFinal.csv")
# get last line
last = all.tail(1)
# clean table
last = last.drop(columns=['Timestamp'])
last = last.drop(columns=['_Sum'])
last = last.dropna(axis='columns')
last
```

```python
values = last.values.tolist()[0]
labels = last.columns.tolist()
```

```python
# draw pie
plt.figure(figsize=(10,10), facecolor='white')
ax1 = plt.subplot()
ax1.pie(values, labels=labels)
plt.show() 
```
